# 使用决策树和随机森林模型进行因子分析

## 决策树和随机森林模型的使用案例

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 生成一个回归任务数据集，样本数量为100
X, y = make_regression(
    n_samples=100, n_features=1, n_targets=1, noise=60, random_state=18
)
plt.scatter(X, y, s=40, c=y, edgecolors="grey", alpha=0.6, cmap='viridis')
plt.colorbar(label='Color Mapping of y')
plt.grid()
plt.show()
```

```python
# 创建决策树实例
reg = DecisionTreeRegressor(max_depth=None)
reg.fit(X, y)
X_new = np.linspace(-2.5, 3.5, 100)
y_new = reg.predict(X_new.reshape(-1, 1))
plt.scatter(X, y, s=40, c=y, edgecolors='grey', alpha=0.4)
plt.plot(X_new, y_new, lw=2, ls='-', c='grey')
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
reg2 = RandomForestRegressor(n_estimators=100)
reg2.fit(X, y)
y_new_2 = reg2.predict(X_new.reshape(-1, 1))
plt.scatter(X, y, s=40, c=y, edgecolors='grey', alpha=0.4)
plt.plot(X_new, y_new, lw=2, ls='-', c='grey', label='Forest')
plt.plot(X_new, y_new_2, lw=2, ls='--', c='r', label='Forest')
plt.legend()
plt.grid()
plt.show()
```
