```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 生产模拟线性回归模拟数据
X, y = make_regression(n_samples=50, n_features=1, noise=40, random_state=88)
plt.scatter(X, y, s=30, c="g", edgecolors="k", alpha=0.6)
plt.grid()
plt.show()
```

```python
print(X[0], y[0])
```

```python
# 训练线性模型
lr = LinearRegression()
lr.fit(X, y)
X_new = np.linspace(-2, 2.5, 100)
y_new = lr.predict(X_new.reshape(-1, 1))
plt.scatter(X, y, s=30, c="g", edgecolors="k", alpha=0.6)
plt.plot(X_new, y_new, c="grey", ls="--")
plt.grid()
plt.show()
```

```python
# 打印模型的斜率和截距
print(lr.coef_, lr.intercept_)
```

```python
pre = lr.predict([[3]])
pre
```

```python
# 正则化线性模型
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=50)
ridge.fit(X, y)
```

```python
y_new2 = ridge.predict(X_new.reshape(-1, 1))
plt.scatter(X, y, s=30, c="g", edgecolors="k", alpha=0.6)
plt.plot(X_new, y_new, c="grey", ls="--", label="Linear")
plt.plot(X_new, y_new2, c="k", ls=":", label="Ridge")
plt.grid()
plt.legend()
plt.show()
```
